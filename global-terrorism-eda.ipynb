```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                     'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                     'nwound': 'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'
                    },inplace=True)
```

```python
# As we have many columns, we take the columns that are necessary for analysis
df = df[['eventid','Year','Month','Day','Country','Region','state','city','latitude','longitude','AttackType','Killed','Wounded',
            'Target','Summary','Group','Target_type','Weapon_type','Motive','success']]
```

```python
df['casualities']=df['Killed']+df['Wounded']
```

```python
df.columns
```

```python
df['state'].fillna('Unknown', inplace=True)
df['city'] = df['city'].fillna(method='ffill')
mean_value=df["latitude"].mean()
df["latitude"].fillna(mean_value,inplace=True)
mean_value=df["longitude"].mean()
df["longitude"].fillna(mean_value,inplace=True)

mean_value=df["Killed"].mean()
df["Killed"].fillna(mean_value,inplace=True)

mean_value=df["Wounded"].mean()
df["Wounded"].fillna(mean_value,inplace=True)
df['Summary'].fillna("Unknow",inplace=True)
df['Motive'].fillna("Unknow",inplace=True)
df['Target'].fillna("Unknow",inplace=True)

mean_value=df["casualities"].mean()
df["casualities"].fillna(mean_value,inplace=True)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(df[: 30], y="state", x="Killed")
plt.show()
```

```python
year_attacks = df.groupby('Year').size().reset_index(name='count')
sns.lineplot(x='Year', y='count', data=year_attacks, color='red')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title("Number of Terrorist Acticity")
plt.show()
```

**There has been a steady increase in global terrorist activities year by year. However, the year 2014 stands out as the peak with the highest recorded incidents. Encouragingly, there has been a subsequent decline in terrorist activity post-2014, offering hope for improved global security efforts.**

# Conclusion

The current state of global affairs presents a troubling escalation in terrorism, casting a shadow over peace and stability worldwide. Notably, regions like the Middle East, North Africa, and South America stand out for the surge in terrorist activities, amplifying the urgency of addressing this pressing issue.

A concerning aspect is the remarkably high success rate of these attacks, which stands at an alarming 89%. This statistic underscores the effectiveness and determination of terrorist entities in executing their destructive agendas, posing significant threats to affected communities.

Of particular concern is the widespread use of bombings and explosions as preferred tactics, resulting in extensive casualties, injuries, and infrastructural damage. The prevalence of such methods underscores the urgent need for comprehensive strategies to combat the proliferation and utilization of explosive devices on a global scale.

As the menace of terrorism persists, tackling its underlying causes, bolstering intelligence capabilities, and fostering international collaboration are indispensable in containing its impact and fostering a safer, more stable world.
